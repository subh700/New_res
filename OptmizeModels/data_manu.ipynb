```python
import pandas as pd
import numpy as np
import glob
from sklearn.model_selection import train_test_split
```

```python
# Load data paths
Seasonailty = glob.glob('Res_Data/*/*/*/Seasonality_test_*.csv')
Trend = glob.glob('Res_Data/*/*/*/Trend_test_*.csv')
Noise = glob.glob('Res_Data/*/*/*/Noise_test_*.csv')
Original = glob.glob('Res_Data/*/*/*/Original_test_*.csv')
Real = glob.glob('Res_Data/*/*/Original.csv')
print(f"Found Seasonailty {len(Seasonailty)} data files")
print(f"Found Trend {len(Trend)} data files")
print(f"Found Noise {len(Noise)} data files")
print(f"Found Original {len(Original)} data files")
print(f"Found Real Data {len(Real)} data files")
# Create initial data list
seas = []
for file_name in Seasonailty:
    df = pd.read_csv(file_name)
    seas.append(df)

Tre = []
for file_name in Trend:
    df = pd.read_csv(file_name)
    Tre.append(df)

Noi = []
for file_name in Noise:
    df = pd.read_csv(file_name)
    Noi.append(df)

Ori = []
for file_name in Original:
    df = pd.read_csv(file_name)
    Ori.append(df)

Rea= []
for file_name in Real:
    df = pd.read_csv(file_name)
    Rea.append(df)
```

```python
for td in range(len(Rea)):
    train_data, test_data = train_test_split(Rea[td], test_size=0.3, shuffle=False)
    # print(f"Training data for index {td}:\n", train_data)
    # print(f"Testing data for index {td}:\n", test_data)
    Test=pd.DataFrame()
    Test['Date']=test_data.iloc[24:,:1].reset_index(drop=True)
    Test['Seasonality_predictions']=seas[td].reset_index(drop=True)
    Test['Trend_predictions']=Tre[td].reset_index(drop=True)
    Test['Noise_predictions']=Noi[td].reset_index(drop=True)
    Test['Original_predictions']=Ori[td].reset_index(drop=True)
    Test['Original_Data']=test_data.iloc[24:,1:2].reset_index(drop=True)
    vi=Seasonailty[td].split('Seasonality')[0]
    Test.to_csv(vi+'Test_pre.csv')
    print(f"Testing data for index {td}:\n", Test)
```
